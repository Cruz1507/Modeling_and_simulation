## Ejemplo 1

metodo de euler con 
\begin{equation}
\frac{dy}{dt} = ln(2)y(t)
\end{equation}

con $y(0)=1$

```python
def step_euler(y,h,f):
  next_y=y+h*f(y)
  return next_y

def euler(h,f,y_0,start_t,end_t):
  N= int((end_t - start_t)/h)
  t_list=np.linspace(start_t,end_t,N+1)
  y_list=np.zeros(N+1)
  y_list[0]=y_0

  for i in range(0,N):
    y_list[i+1]=step_euler(y_list[i],h,f)
  return t_list,y_list
```

```python
def g(y):
  crecimiento=np.log(2)*y
  return crecimiento


y_0=1
h=0.01
t_0=0
t_N=10

t_list,y_list=euler(h,g,y_0,t_0,t_N)

plt.plot(t_list,y_list)
plt.plot(t_list, np.power(2,t_list))
plt.grid()
```

Cuando $a$ es positivo obtenemos 3 puntos de equilibrio 
$$x=0$$
$$x=\sqrt{a}$$
$$x=-\sqrt{a}$$
Donde las estabilidades estaran en las raices de a

```python
a=9
def g(y):
  crecimiento = a*y-y**3
  return crecimiento

x=np.linspace(-5,5)
plt.plot(x,g(x))
plt.grid()
```
