```python
import numpy as np
import matplotlib.pyplot as plt
import math
import sympy as sp
from scipy.integrate import solve_ivp
from ipywidgets import interact
```

### 6.4.1
\begin{equation}
  \dot{x}=x(3-x-y), \dot{y}=y(2-x-y)
\end{equation}

```python
from sympy import var, solve
x, y = var('x y')

f1 = -(x**2)+3*x-x*y
f2 = -(y**2)+2*y-x*y

fig, ax = plt.subplots(figsize=(12,8))

sols = solve((f1, f2), (x, y))
for r in sols:
    plt.plot(r[0],r[1],'ro')

X,Y=np.meshgrid(np.linspace(0,4,40),np.linspace(0,3,30))
V= -(X**2)+3*X-X*Y
U= -(Y**2)+2*Y-X*Y
plt.title("Grafica plano de fase", 
          fontdict={'color' : 'darkblue',
                    'size': 18})
plt.streamplot(X,Y,V, U, density=1., arrowsize=1)
plt.grid()
plt.plot()
```
